```python
# We can access all the pandas functions with the alias pd using import
import pandas as pd 
```

```python
# *** DATA COLLECTION *** #

dataset=pd.read_csv("Salary_Data.csv") # Read the scv file (i.e Dataset)
dataset # To view the output of dataset(YearsExperience,Salary)
```

```python
# *** INPUT *** #

independent=dataset[["YearsExperience"]] # Input is assigned to independent variable
independent # To view the output of independent variable(YearsExperience)
```

```python
# *** OUTPUT *** #

dependent=dataset[["Salary"]] # Ouput is assigned to dependent variable
dependent # To view the output of dependent variable(Salary)
```

```python
# *** SPLIT TRAIN and TEST *** #

from sklearn.model_selection import train_test_split #import splitting dataset(train and test set) function from sklearn library

# assignning input(X) and output(Y) variable for training set(X_train,X_test) and test set(Y_train,Y_test)
# 30% of data will set for testing and 70% is used for training
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# *** LINEAR REGRESSION *** # 

from sklearn.linear_model import LinearRegression #import LinearRegression function from sklearn library (fitting a straignt line to data)
regressor=LinearRegression() # assigned linearregression class into regressor variable
regressor.fit(X_train,Y_train) #trains the linear regression model by finding the best-fit line between the input (X_train) and output(Y_train).
```

```python
weight=regressor.coef_ #weight is calculated in coef and assigned the values in weight variable
weight # calling the variable to display its value
```

```python
bais=regressor.intercept_ #bais is calculated in intercept and assigned the values(slope) in bais variable
bais # calling the variable to display its value
```

```python
# *** EVALUTION *** # 

y_pred=regressor.predict(X_test) #crosss checking (i.e compare the predicted(y_pred) value with actual value)
```

```python
# *** R2  *** # 

from sklearn.metrics import r2_score #import r2_score function from sklearn library (Measures how well the regression line fits the data (1 = perfect fit, 0 = no fit).
r_score=r2_score(Y_test,y_pred) # Passing input
r_score 
```

```python
# *** SAVE THE MODEL *** # 

import pickle
filename="finalized_model_linear.sav" #name of the file where the model will saved
pickle.dump(regressor,open(filename,'wb')) # train the linear reegression model and open the file in write binary mode, also write the serialised version of regrsssor into the file)
```
